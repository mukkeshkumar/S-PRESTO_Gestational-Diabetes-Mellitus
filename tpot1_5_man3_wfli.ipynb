Generating X & Y matrices

```python
X=df2.loc[:, df2.columns != 'PGV4_OGTT_IADPSG2018_strict']
y=df2.loc[:, df2.columns == 'PGV4_OGTT_IADPSG2018_strict']
```

```python
X.columns
```

Define model evaluation

```python
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.10, random_state=0)
```

Define TPOT search

```python
model = TPOTClassifier(generations=100, population_size=100, scoring='roc_auc', cv=cv, n_jobs=50, random_state=0, config_dict='TPOT NN', verbosity=2)
```

Perform TPOT search

```python
model.fit(X, y)
```

Export the best model

```python
model.export('tpot_best_model.py')
```
